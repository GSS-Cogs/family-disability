```python
figure_a_filter = 'http://gss-data.org.uk/dataset/download?uri=http%3A%2F%2Fgss-data.org.uk%2Fdata%2Fgss_data%2Fdisability%2Fdfe-special-educational-needs&filter-facet=http%253A%252F%252Fgss-data.org.uk%252Fdef%252Fdimension%252Fdfe-special-education-provider%2Chttp%253A%252F%252Fgss-data.org.uk%252Fdef%252Fconcept%252Fdfe-special-education-provider%252Fall-schools&filter-facet=http%253A%252F%252Fgss-data.org.uk%252Fdef%252Fdimension%252Fdfe-special-education-need-type%2Chttp%253A%252F%252Fgss-data.org.uk%252Fdef%252Fconcept%252Fdfe-special-education-need-type%252Fpupils-on-roll&filter-facet=http%253A%252F%252Fgss-data.org.uk%252Fdef%252Fdimension%252Fdfe-special-education-need-type%2Chttp%253A%252F%252Fgss-data.org.uk%252Fdef%252Fconcept%252Fdfe-special-education-need-type%252Fpupils-with-sen&filter-facet=http%253A%252F%252Fgss-data.org.uk%252Fdef%252Fdimension%252Fdfe-special-education-need-type%2Chttp%253A%252F%252Fgss-data.org.uk%252Fdef%252Fconcept%252Fdfe-special-education-need-type%252Fpupils-with-sen-without-statements-or-ehc-plans&filter-facet=http%253A%252F%252Fgss-data.org.uk%252Fdef%252Fdimension%252Fdfe-special-education-need-type%2Chttp%253A%252F%252Fgss-data.org.uk%252Fdef%252Fconcept%252Fdfe-special-education-need-type%252Fpupils-with-statements-or-ehc-plans&filter-facet=http%253A%252F%252Fgss-data.org.uk%252Fdef%252Fdimension%252Fdfe-special-education-support-type%2Chttp%253A%252F%252Fgss-data.org.uk%252Fdef%252Fconcept%252Fdfe-special-education-support-type%252Fprovision'
table_a = pd.read_csv(figure_a_filter)
```

In order to derieve the % for each Education Need Type I will have to extract 'Pupils On Roll' value for each reference period and from there calculate the percentage. 
- by going through each row and checking if it the associated values match and storing that total. 

```python
pupils_on_roll_7 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2007)) , "Value"] 
pupils_on_roll_8 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2008)) , "Value"] 
pupils_on_roll_9 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2009)) , "Value"] 
pupils_on_roll_10 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2010)) , "Value"] 
pupils_on_roll_11 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2011)) , "Value"] 
pupils_on_roll_12 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2012)) , "Value"] 
pupils_on_roll_13 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2013)) , "Value"] 
pupils_on_roll_14 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2014)) , "Value"] 
pupils_on_roll_15 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2015)) , "Value"] 
pupils_on_roll_16 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2016)) , "Value"] 
pupils_on_roll_17 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2017)) , "Value"] 
pupils_on_roll_18 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2018)) , "Value"] 
pupils_on_roll_19 = table_a.loc[((table_a['DfE Special Education Need Type'] == 'Pupils On Roll') & (table_a["Reference Period"]==2019)) , "Value"] 
```

```python
table_a['Total'] = ''
table_a.loc[table_a['Reference Period'].eq(2007), 'Total'] = pupils_on_roll_7.values[0]
table_a.loc[table_a['Reference Period'].eq(2008), 'Total'] = pupils_on_roll_8.values[0]
table_a.loc[table_a['Reference Period'].eq(2009), 'Total'] = pupils_on_roll_9.values[0]
table_a.loc[table_a['Reference Period'].eq(2010), 'Total'] = pupils_on_roll_10.values[0]
table_a.loc[table_a['Reference Period'].eq(2011), 'Total'] = pupils_on_roll_11.values[0]
table_a.loc[table_a['Reference Period'].eq(2012), 'Total'] = pupils_on_roll_12.values[0]
table_a.loc[table_a['Reference Period'].eq(2013), 'Total'] = pupils_on_roll_13.values[0]
table_a.loc[table_a['Reference Period'].eq(2014), 'Total'] = pupils_on_roll_14.values[0]
table_a.loc[table_a['Reference Period'].eq(2015), 'Total'] = pupils_on_roll_15.values[0]
table_a.loc[table_a['Reference Period'].eq(2016), 'Total'] = pupils_on_roll_16.values[0]
table_a.loc[table_a['Reference Period'].eq(2017), 'Total'] = pupils_on_roll_17.values[0]
table_a.loc[table_a['Reference Period'].eq(2018), 'Total'] = pupils_on_roll_18.values[0]
table_a.loc[table_a['Reference Period'].eq(2019), 'Total'] = pupils_on_roll_19.values[0]

table_a.drop(table_a.loc[table_a['DfE Special Education Support Type']=='Total'].index, inplace=True)
```

```python
#calculate percentage
table_a['%'] = (table_a['Value'] / table_a['Total'] * 100) 
table_a['%'] = table_a['%'].round(1)
```

```python
#drop what is no longer needed
table_a.drop(table_a.loc[table_a['%']==100].index, inplace=True)
table_a = table_a.drop(['DfE Special Education Support Type', 'DfE Age Groups', 'DfE Special Education Provider', 
              'measure type', 'Value', 'Unit of Measure', 'Total', 'DfE Sex Groups', 'Reference Area'], axis=1)
```

```python
#piviot table
new_table_a = pd.pivot_table(table_a,index=["Reference Period"], columns='DfE Special Education Need Type', values='%')
new_table_a
```

```python
#Plot line graph 
new_table_a.plot(figsize=(16, 9), title='Figure A:');
plt.ticklabel_format(style='plain', axis='y')
plt.ylabel("%")
```
